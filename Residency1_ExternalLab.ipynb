### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
male=0
female=0
for item in HR_external["Gender"]:
    if item==1:
        male+=1
    else:
        female+=1
print("Number of Male Employees=",male,"\nNumber of Female Employees=",female)
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
q1=HR_external['Age'].quantile(0.25)
q3=HR_external['Age'].quantile(0.75)
iqr=q3-q1
print("1st Quantile=",q1,"\n3rd Quantile=",q3,"\nIQR=",iqr)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
print("Mean Value of DailyRate=",HR_external["DailyRate"].mean())
print("Median Value of DailyRate=",HR_external["DailyRate"].median())
print("Variance Value of DailyRate=",HR_external["DailyRate"].median())
```
