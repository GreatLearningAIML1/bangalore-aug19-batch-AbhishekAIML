```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.(5 points)
import pandas as pd
#Read Laliga csv into laliga DataFrame
laliga=pd.read_csv('Laliga.csv')
#Assign 1st row as column name
laliga=laliga.rename(columns=laliga.loc[0]).drop(laliga.index[0])
#Replace all dashes with 0
laliga.loc[:,'Points':'T'].replace({'-':0},regex=True,inplace=True)
#convert to int before applying as row index
#laliga.loc[:,'Pos']=laliga.loc[:,'Pos'].astype(int)
laliga['Pos']=pd.to_numeric(laliga['Pos'])
#Apply as row index
laliga.set_index('Pos',inplace=True)
#convert to int so that Arithmetic operations can be performed
#laliga.loc[:,'Seasons':'T'] = laliga.loc[:,'Seasons':'T'].astype(int)
laliga['Team']=pd.Categorical(laliga['Team'])
laliga['Seasons']=pd.to_numeric(laliga['Seasons'])
laliga['Points']=pd.to_numeric(laliga['Points'])
laliga['GamesPlayed']=pd.to_numeric(laliga['GamesPlayed'])
laliga['GamesWon']=pd.to_numeric(laliga['GamesWon'])
laliga['GamesDrawn']=pd.to_numeric(laliga['GamesDrawn'])
laliga['GamesLost']=pd.to_numeric(laliga['GamesLost'])
laliga['GoalsFor']=pd.to_numeric(laliga['GoalsFor'])
laliga['GoalsAgainst']=pd.to_numeric(laliga['GoalsAgainst'])
laliga['Champion']=pd.to_numeric(laliga['Champion'])
laliga['Runner-up']=pd.to_numeric(laliga['Runner-up'])
laliga['Third']=pd.to_numeric(laliga['Third'])
laliga['Fourth']=pd.to_numeric(laliga['Fourth'])
laliga['Fifth']=pd.to_numeric(laliga['Fifth'])
laliga['Sixth']=pd.to_numeric(laliga['Sixth'])
laliga['T']=pd.to_numeric(laliga['T'])
laliga['Debut']=pd.Categorical(laliga['Debut'])
laliga['Since/LastApp']=pd.Categorical(laliga['Since/LastApp'])
laliga['BestPosition']=pd.to_numeric(laliga['BestPosition'])
#info
print(laliga.info())
#Random Arithmetic operation test
print('Sum of Champion column is = ',laliga['Champion'].sum())
laliga.head()
```

```python
#Print all the teams which have started playing between 1930-1980. (5 points)
laliga[pd.to_numeric(laliga.Debut.str.split('-').str[0]).between(1930, 1981)].loc[:,('Team','Debut')]
```

```python
#Print the list of teams which came Top 5 in terms of points (5 points)
laliga.sort_values(by=['Points','Team'],ascending=False,inplace=False).loc[1:5,('Team','Points')]
```

```python
""" Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst """
def Goal_diff_count(df):
    laliga['Goal_diff_count']=laliga.GoalsFor-laliga.GoalsAgainst
    min_diff=laliga[laliga['Goal_diff_count']==laliga['Goal_diff_count'].min()]
    max_diff=laliga[laliga['Goal_diff_count']==laliga['Goal_diff_count'].max()]
    return min_diff.loc[:,('Team','Goal_diff_count')], max_diff.loc[:,('Team','Goal_diff_count')], laliga.loc[:,('Team','Goal_diff_count')]
min_diff,max_diff, all_team_goal_diff=Goal_diff_count(laliga)
print('Team which has minimum goal difference -\n\n',min_diff)
print('Team which has maximum goal difference -\n\n',max_diff)
print('All team goal difference -\n\n',all_team_goal_diff)
```

```python
'''
Create a new column with name “Winning Percent” and append it to the data set
Percentage of Winning = (GamesWon / GamesPlayed)*100
If there are any numerical error, replace it with 0%
Print the top 5 teams which has the highest Winning percentage (5 points)
'''
laliga['Winning Percent']=(laliga['GamesWon']/laliga['GamesPlayed']*100).fillna(0)
print('Top 5 teams which has the highest Winning percentage \n',laliga.sort_values(by=['Winning Percent','Team'],ascending=False,inplace=False).loc[1:5,('Team','Winning Percent')])
laliga.tail(5)
```

```python
'''
Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
Eg: Best Position                Points
        1                        25000
        2                        7000 
'''
temp=laliga.groupby(['BestPosition']).sum().loc[:,'Points']
print('Sum of the Points based on Best Position -')
print('Best Position\t' 'Points')
for label, content in temp.items():
    print(label,content, sep='\t\t')
```
